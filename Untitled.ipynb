```julia
risk_free_rate = 0.05;
```

```julia
all_tickers = keys(dataset) |> collect |> sort;
```

```julia
all_firms_excess_return_matrix = log_return_matrix(dataset, all_tickers, 
    Δt = (1.0/252.0), risk_free_rate = risk_free_rate);
```

```julia
μ = mean(all_firms_excess_return_matrix, dims=1) |> vec;
```

```julia
sims = load(joinpath(_PATH_TO_DATA, "SIMs-Kuang-PD1-CHEME-5660-Fall-2023.jld2")) |> x->x["sims"];
```

```julia
index_SPY = findfirst(x->x=="SPY", all_tickers);
R_SPY = μ[index_SPY]
```

```julia
μ_sim2 = Array{Float64,1}();
for i ∈ eachindex(all_tickers)
    
    myticker = all_tickers[i];
    sim = sims[myticker];
    
    αᵢ = sim.α
    βᵢ = sim.β
    Rᵢ = αᵢ+βᵢ*R_SPY
    
    push!(μ_sim2,Rᵢ)
end
```

```julia
σₘ = std(all_firms_excess_return_matrix[:, index_SPY]);
```

```julia
Σ_tmp = Array{Float64,2}(undef, length(μ), length(μ));
for i ∈ eachindex(all_tickers)
    outer_ticker = all_tickers[i];
    sim_outer = sims[outer_ticker];
    
    for j ∈ eachindex(all_tickers)
        
        inner_ticker = all_tickers[j];
        sim_inner = sims[inner_ticker];
        
        if (i == j)
            βᵢ = sim_outer.β
            ϵᵢ = sim_outer.ϵ
            σ_ϵᵢ = params(ϵᵢ)[2];
            Σ_tmp[i,j] = ((βᵢ)^2)*((σₘ)^2)+(σ_ϵᵢ)^2
        else
            βᵢ = sim_outer.β
            βⱼ = sim_inner.β
            Σ_tmp[i,j] = βᵢ*βⱼ*(σₘ)^2
        end
    end
end
Σ_sim2  = Σ_tmp |> x-> x*(1/252);
```

```julia

my_list_of_tickers2 = ["SPY"];
my_list_of_firm_ids2 = Array{Int64,1}();
for ticker ∈ my_list_of_tickers2
    firm_index = findfirst(x->x==ticker, all_tickers);    
    push!(my_list_of_firm_ids2, firm_index)
end
```

```julia
# --- DO NOT CHANGE THIS BLOCK ----------------------------------------- #
μ̂_sim2 = Array{Float64,1}();
for firm_index ∈ my_list_of_firm_ids2
    push!(μ̂_sim2, μ_sim2[firm_index])
end
# ---------------------------------------------------------------------- #
```

```julia
# --- DO NOT CHANGE THIS BLOCK FOR SIM  -------------------------------- #
my_number_of_selected_firms2 = length(my_list_of_firm_ids2)
Σ̂_sim2 = Array{Float64,2}(undef, my_number_of_selected_firms2, my_number_of_selected_firms2);
for i ∈ eachindex(my_list_of_firm_ids2)
    row_firm_index = my_list_of_firm_ids2[i]
    for j ∈ eachindex(my_list_of_firm_ids2)
        col_firm_index = my_list_of_firm_ids2[j]
        Σ̂_sim2[i,j] = Σ_sim2[row_firm_index, col_firm_index]
    end
end
# ---------------------------------------------------------------------- #
```

```julia
my_data2 = Dict{String,Any}();
my_data2["expected_excess_return"] = μ̂_sim2;
my_data2["covariance"] = Σ̂_sim2;
save(joinpath(_PATH_TO_DATA,"MyChoiceSet-Example2.jld2"), Dict("dataset" => my_data2));
```

```julia
number_of_firms2 = length(my_list_of_tickers2);
wₒ = zeros(number_of_firms2);
wₒ[1] = 1.0;
bounds = zeros(number_of_firms2,2);
bounds[:,2] .= 1.0;
number_of_points = 40;
```

```julia
problem_risk_sim2 = build(MyMarkowitzRiskyAssetOnlyPortfiolioChoiceProblem, (
    Σ = Σ̂_sim2,
    μ = μ̂_sim2,
    bounds = bounds,
    initial = wₒ,
    R = 0.0
));
```

```julia
minimum_desired_reward_array2 = range(0.0, stop = 0.5 - risk_free_rate, length = number_of_points) |> collect;
```

```julia
efficient_frontier_sim2 = Dict{Float64,Float64}();
portfolio_df = DataFrame();
for i ∈ eachindex(minimum_desired_reward_array2)
    
    # update the problem object -
    problem_risk_sim2.R = minimum_desired_reward_array2[i];
    
    # compute -
    solution_sim2 = solve(problem_risk_sim2)

    # check: did this converge?
    status_flag = solution_sim2["status"];    
    if (status_flag == MathOptInterface.LOCALLY_SOLVED)
        key = sqrt(solution_sim2["objective_value"]);
        value = solution_sim2["reward"];
        efficient_frontier_sim2[key] = value;
        
        w_opt = solution_sim2["argmax"];
        
        # add data to portfolio_df -
        row_df = (
            expected_excess_return = value,
            risk = key,
            tickers = my_list_of_tickers2,
            w = w_opt,
            risk_free_rate = risk_free_rate
        )
        push!(portfolio_df,row_df);
    end
end
efficient_frontier_sim2
```

```julia
# single index model -
plot(efficient_frontier_sim2, lw=4, xlabel="Portfolio Risk (sqrt)", 
    ylabel="Portfolio Excess Return", fontsize=18, label="SIM", c=:red, 
    xlim=(0.0, 1.2*maximum(efficient_frontier_sim2).first))
scatter!(efficient_frontier_sim2, label="", c=:white, mec=:red, ms=3)
```

```julia
Wₒ = 24000000;
number_of_trading_days = 3650;
offset = 1;
```

```julia
P = rand(efficient_frontier_sim2);
σₚ = P.first
μₚ = P.second
```

```julia
wealth_array2 = Array{Float64,2}(undef, number_of_trading_days, 8);
for i ∈ 1:number_of_trading_days
    Δt = (i-1)*(1/252);
    
    wealth_array2[i,1] = Δt
    wealth_array2[i,2] = Wₒ*exp((μₚ + risk_free_rate)*Δt);
    wealth_array2[i,3] = Wₒ*exp((μₚ - σₚ + risk_free_rate)*Δt);
    wealth_array2[i,4] = Wₒ*exp((μₚ + σₚ + risk_free_rate)*Δt);
    wealth_array2[i,5] = Wₒ*exp((μₚ - 1.96*σₚ + risk_free_rate)*Δt);
    wealth_array2[i,6] = Wₒ*exp((μₚ + 1.96*σₚ + risk_free_rate)*Δt);
    wealth_array2[i,7] = Wₒ*exp((μₚ - 2.576*σₚ + risk_free_rate)*Δt);
    wealth_array2[i,8] = Wₒ*exp((μₚ + 2.576*σₚ + risk_free_rate)*Δt);
end
```
